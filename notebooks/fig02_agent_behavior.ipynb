```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from functions import generate_D, generate_F
```

```python
resolution = 10
radius = resolution 
total_erosion = np.load('fig02/erosion_demo_total_erosion.npy')[75-40:75+40,:]
resource = np.load('fig02/erosion_demo_resource.npy')[75-40:75+40,:]
line = np.load('fig02/erosion_demo_line.npy')

# Create the 3D surface plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d', proj_type='ortho', computed_zorder=False)

# Create a grid for the surface plot
x = np.arange(total_erosion.shape[1])
y = np.arange(total_erosion.shape[0])
x, y = np.meshgrid(x, y)

# Normalize the difference for coloring
norm = plt.Normalize(total_erosion.min(), total_erosion.max())
colors = plt.cm.jet(norm(total_erosion))

# Plot the erosion
ax.plot_surface(x, y, total_erosion*-50, facecolors=colors, cstride=1, rstride=1, zorder=0)

# Plot the resource
norm = plt.Normalize(resource.min(), resource.max())
colors = plt.cm.Greens(norm(resource))
ax.plot_surface(x, y, resource+70, facecolors=colors, cstride=1, rstride=1, zorder=-1)
ax.scatter(line[1:,0], line[1:,1]-35, np.ones(np.shape(line[1:,0]))+70, zorder=1, c='k', alpha=1, s=10)

# Plot the path
ax.plot(line[:,0], line[:,1]-35, np.ones(np.shape(line[:,0]))+140, zorder=2, c='k', lw=2)
ax.scatter(line[1:-1,0], line[1:-1,1]-35, np.ones(np.shape(line[1:-1,0]))+140, zorder=2, c='k', alpha=1, s=40)
ax.scatter(line[-1,0], line[-1,1]-35, np.ones(1)+140, marker=(3, 0, 25), zorder=2, c='k', alpha=1, s=150)

tempx = [0, 150, 150, 0, 0]
tempy = [0, 0, 80, 80, 0]
ax.plot(tempx, tempy, np.ones(np.shape(tempx))+140, zorder=2, c='k', alpha=0.5)

# Remove axes lines and ticks
ax.axis('off')
ax.view_init(elev=30, azim=-80)#, elev=0)
ax.set_aspect('equal')

ax.grid(False)
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

plt.show()

# First plot: with opaque brush
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
brush = generate_D(2 * radius)
ax.imshow(brush, cmap='binary')
ax.set_xlim((-0.5*resolution, 4.5*resolution))
ax.set_ylim((-0.5*resolution, 4.5*resolution))

# Remove tick marks and labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])

plt.show()

# Second plot: with spherical brush
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
brush = generate_F(2 * radius)
ax.imshow(brush, cmap='binary')
ax.set_xlim((-0.5*resolution, 4.5*resolution))
ax.set_ylim((-0.5*resolution, 4.5*resolution))

# Remove tick marks and labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])

plt.show()


```
